Untuk mengganti *missing values* pada setiap kolom perlu dilakukan visualisasi data melalui box plot atau hist plot untuk menentukan data tersebut harus diganti dengan mean atau median.

```python
for i in df.columns:
    fig, axs = plt.subplots(1,2,figsize=(15, 3))

    sns.histplot(df[i],bins=20, kde=True,ax=axs[0])
    sns.boxplot(data = df[i], x = df[i], ax = axs[1], fliersize=1)
```

Berdasarkan visualisasi sebaran data, kolom 'Rainfall', 'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm' karena terlalu banyak outliers sehingga perlu dilkukan *handling outliers* agar tidak mengacaukan data.

```python
# Handling outlier menggunakan IQR

# Menghitung Q1, Q3, dan IQR
Q1 = df[['BALANCE', 'BALANCE_FREQUENCY', 'ONEOFF_PURCHASES_FREQUENCY', 
         'PURCHASES_INSTALLMENTS_FREQUENCY', 'CASH_ADVANCE_FREQUENCY', 
         'PURCHASES_TRX', 'CREDIT_LIMIT', 
         'PAYMENTS', 'PRC_FULL_PAYMENT']].quantile(0.25)
Q3 = df[['BALANCE', 'BALANCE_FREQUENCY', 'ONEOFF_PURCHASES_FREQUENCY', 
         'PURCHASES_INSTALLMENTS_FREQUENCY', 'CASH_ADVANCE_FREQUENCY', 
         'PURCHASES_TRX', 'CREDIT_LIMIT', 
         'PAYMENTS', 'PRC_FULL_PAYMENT']].quantile(0.75)
IQR = Q3 - Q1

# Menghitung batas atas (maximum) dan batas bawah (minimum)
maximum = Q3 + 1.5 * IQR
minimum = Q1 - 1.5 * IQR

print('Nilai Maximum : \n', maximum)
print('Nilai Minimum : \n', minimum)
```

```python
# Melakukan align pada DataFrames
df, maximum = df.align(maximum, axis=1, join='outer')
df, minimum = df.align(minimum, axis=1, join='outer')

more_than = (df > maximum)
lower_than = (df < minimum)

# Menerapkan mask pada axis=1 (kolom)
df = df.mask(more_than, maximum, axis=1)
df = df.mask(lower_than, minimum, axis=1)
```

```python
rows = 6
cols = 3

# Increase figure size and resolution
plt.figure(figsize=(23, 28), dpi=100)

for i, column in enumerate(df.columns, 1):
    plt.subplot(rows, cols, i)
    fig = sns.boxplot(data=df[column], x=df[column], fliersize=1)
    fig.set_title('')
    fig.set_ylabel(column)

plt.tight_layout()
plt.show()
```

Sekarang outlier pada semua kolom sudah dihilangkan, kemudian untuk mengisi NaN pada kolom MINIMUM_PAYMENTS dan kolom CREDIT_LIMIT maka perlu dilihat dengan lebih detail persebaran data pada kolom tersebut

```python
missing = ["MINIMUM_PAYMENTS", "CREDIT_LIMIT"]

for i in missing:
    fig, axs = plt.subplots(1, 2, figsize=(15, 3))

    sns.histplot(df[i], bins=20, kde=True, ax=axs[0])
    sns.boxplot(data=df[i], x=df[i], ax=axs[1], fliersize=1)

    axs[0].set_title(f'Histogram of {i}')
    axs[1].set_title(f'Boxplot of {i}')

plt.show()
```

Karena persebaran data memiliki skewnes, missing values digantikan dengan median data

```python
# Mengganti missing values dengan median
df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].median(), inplace=True)
df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].median(), inplace=True)
```

#### 6.2 Pendefinisian Model Clustering

```python
inertia = []

for i in range(1, 10):
    model = KMeans(n_clusters=i, random_state=42)
    model.fit(normalised)
    inertia.append(model.inertia_)

inertia_df = pd.DataFrame({'Inertia':inertia}, index=[i for i in range(1, 10)])
inertia_df
```

```python
inertia_df.plot(kind='line')
plt.ylabel('Inertia (Sum of Squared Error)')
plt.xlabel('Number of Clusters')
plt.show()
```

<b> Insight : </b> Nilai inertia menunjukkan bahwa k-kluster yang dapat dipilih yakni 4-6 kluster. Dalam memilih banyaknya kluster harus mempertimbangkan computation cost, sehingga dipilih banyaknya k kluster yakni 4 kluster.

```python
# Model K-means
km = KMeans(n_clusters=4, random_state=42)

# Model Hierachical Clustering
agg = AgglomerativeClustering(n_clusters = 4, affinity = "euclidean", linkage = "ward")
```

#### 6.3 Pelatihan Model

```python
# Model Hierachical Clustering
normalised['Cluster AGG'] = agg.fit_predict(normalised)
```

#### 6.5 Evaluasi Model

```python
normalised.iloc[:,0:14].head()
```

```python
sil_km = silhouette_score(normalised.iloc[:,0:14], normalised['Cluster KM'])
sil_agg = silhouette_score(normalised.iloc[:,0:14], normalised['Cluster AGG'])

ch_km = calinski_harabasz_score(normalised.iloc[:,0:14], normalised['Cluster KM'])
ch_agg = calinski_harabasz_score(normalised.iloc[:,0:14], normalised['Cluster AGG'])

db_km = davies_bouldin_score(normalised.iloc[:,0:14], normalised['Cluster KM'])
db_agg = davies_bouldin_score(normalised.iloc[:,0:14], normalised['Cluster AGG'])
```

```python
model_selection = pd.DataFrame({'KMeans': [sil_km, ch_km, db_km],
                                'AgglomerativeClustering': [sil_agg, ch_agg, db_agg]},
                               index = ['Silhouette Score', 'Calinski Harabasz', 'Davies-Bouldin'])
model_selection
```

<b> Insight : </b>
