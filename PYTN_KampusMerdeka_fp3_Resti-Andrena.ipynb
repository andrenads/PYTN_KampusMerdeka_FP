## Data Preprocessing

### Label Encoding

Dilakukan encoding untuk mengubah keseluruhan data menjadi data numerik.

```python
le = LabelEncoder()
columns_to_encode = ['anaemia', 'diabetes', 'high_blood_pressure', 'sex', 'smoking', 'DEATH_EVENT']
for col in columns_to_encode:
    heart_failure[col] = le.fit_transform(heart_failure[col])
```

```python
heart_failure.head()
```

```python
correlation = heart_failure.corr()
plt.figure(figsize=(16,12))
plt.title('Correlation Heatmap of Heart Failure')
ax = sns.heatmap(correlation, square=True, annot=True, fmt='.1f', linecolor='white')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_yticklabels(ax.get_yticklabels(), rotation=30)           
plt.show()
```

Berdasarkan output heatmap, dapat diambil kesimpulan bahwa kolom _age_, _serum_creatinine_, _time_, _high blood pressure_, dan _serum_sodium_ memiliki korelasi yang tinggi dengan DEATH_EVENT.

### Splitting dataset

```python
X = heart_failure.drop("DEATH_EVENT", axis=1)
y = heart_failure.DEATH_EVENT
```

```python
X.head()
```

```python
y.head()
```

```python
random_state = 50
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=random_state)

print(f"Training dataset\t: {len(X_train)}\t({round(len(X_train) / len(X) * 100)}%)")
print(f"Testing dataset\t\t: {len(X_test)}\t({round(len(X_test) / len(X) * 100)}%)")
```

```python
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

### Defining Model

```python
rf_model = RandomForestClassifier(n_estimators=100, max_depth=10)
```

```python
# Boosting Method
gbm = GradientBoostingClassifier(learning_rate=0.01, random_state=1)
lr_boosting = AdaBoostClassifier(LogisticRegression(C=10000, solver='lbfgs'), n_estimators=100, learning_rate = 0.1)
svm_boosting = AdaBoostClassifier(SVC(C=100000, gamma='scale', probability=True), n_estimators=100, learning_rate = 0.1)
```

### Model Training & Model Evaluation

```python
# Melatih dan mengevaluasi Random Forest
rf_model.fit(scaled_X_train, y_train)
y_pred_rf = rf_model.predict(scaled_X_test)

# Membuat Confusion Matrix sebagai DataFrame
cm = confusion_matrix(y_test, y_pred_rf)
conf_matrix_df = pd.DataFrame(cm, index=['Actual 0', 'Actual 1'], columns=['Predicted 0', 'Predicted 1'])

# Menampilkan Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred_rf))
print("======================================================\n")

# Menampilkan Tabel Confusion Matrix
print("Model: Random Forest")
print("Confusion Matrix:")
print(conf_matrix_df)

# Menampilkan Tabel Confusion Matrix dengan Heatmap (seaborn)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', annot_kws={"size": 16})
plt.title("Confusion Matrix - Random Forest")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

print(f'Accuracy Score: {accuracy_score(y_test, y_pred_rf):.3f}')
print(f'Precision Score: {precision_score(y_test,y_pred_rf):.3f}')
print(f'Recall Score: {recall_score(y_test,y_pred_rf):.3f}')
print("F1 Score:{:.3f}".format(f1_score(y_test,y_pred_rf)))
```

```python
# Gradient Boosting 
gbm.fit(scaled_X_train, y_train)
y_pred_gbm = gbm.predict(scaled_X_test)

# Membuat Confusion Matrix sebagai DataFrame
cm = confusion_matrix(y_test, y_pred_gbm)
conf_matrix_df = pd.DataFrame(cm, index=['Actual 0', 'Actual 1'], columns=['Predicted 0', 'Predicted 1'])

# Menampilkan Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred_gbm))
print("======================================================\n")

# Menampilkan Tabel Confusion Matrix
print("Model: Gradient Boosting")
print("Confusion Matrix:")
print(conf_matrix_df)

# Menampilkan Tabel Confusion Matrix dengan Heatmap (seaborn)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', annot_kws={"size": 16})
plt.title("Confusion Matrix - Gradient Boosting")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

print(f'Accuracy Score: {accuracy_score(y_test, y_pred_gbm):.3f}')
print(f'Precision Score: {precision_score(y_test,y_pred_gbm):.3f}')
print(f'Recall Score: {recall_score(y_test,y_pred_gbm):.3f}')
print("F1 Score:{:.3f}".format(f1_score(y_test,y_pred_gbm)))
```

```python
# Logistic Regression
lr_boosting.fit(scaled_X_train, y_train)
y_pred_lr = lr_boosting.predict(scaled_X_test)

# Membuat Confusion Matrix sebagai DataFrame
cm = confusion_matrix(y_test, y_pred_lr)
conf_matrix_df = pd.DataFrame(cm, index=['Actual 0', 'Actual 1'], columns=['Predicted 0', 'Predicted 1'])

# Menampilkan Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred_lr))
print("======================================================\n")

# Menampilkan Tabel Confusion Matrix
print("Model: Logistic Regresion")
print("Confusion Matrix:")
print(conf_matrix_df)

# Menampilkan Tabel Confusion Matrix dengan Heatmap (seaborn)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', annot_kws={"size": 16})
plt.title("Confusion Matrix - Logistic Regresion")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

print(f'Accuracy Score: {accuracy_score(y_test, y_pred_lr):.3f}')
print(f'Precision Score: {precision_score(y_test,y_pred_lr):.3f}')
print(f'Recall Score: {recall_score(y_test,y_pred_lr):.3f}')
print("F1 Score:{:.3f}".format(f1_score(y_test,y_pred_lr)))
```

```python
# SVM
svm_boosting.fit(scaled_X_train, y_train)
y_pred_svm = svm_boosting.predict(scaled_X_test)

# Membuat Confusion Matrix sebagai DataFrame
cm = confusion_matrix(y_test, y_pred_svm)
conf_matrix_df = pd.DataFrame(cm, index=['Actual 0', 'Actual 1'], columns=['Predicted 0', 'Predicted 1'])

# Menampilkan Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred_svm))
print("======================================================\n")

# Menampilkan Tabel Confusion Matrix
print("Model: SVM")
print("Confusion Matrix:")
print(conf_matrix_df)

# Menampilkan Tabel Confusion Matrix dengan Heatmap (seaborn)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', annot_kws={"size": 16})
plt.title("Confusion Matrix - SVM")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

print(f'Accuracy Score: {accuracy_score(y_test, y_pred_svm):.3f}')
print(f'Precision Score: {precision_score(y_test,y_pred_svm):.3f}')
print(f'Recall Score: {recall_score(y_test,y_pred_svm):.3f}')
print("F1 Score:{:.3f}".format(f1_score(y_test,y_pred_svm)))
```

Setelah dilakukan seluruh pengujian, nilai akurasi tertinggi terjadi pada model menggunakan Random Forest yaitu 90,6% sedangakan model lain dengan menggunakan Gradient Boosting (84,9%), Logistic Regression (75,5%), dan SVM (71,7%). Maka dipilih model dengan menggunakan Random Forest Classifier.
